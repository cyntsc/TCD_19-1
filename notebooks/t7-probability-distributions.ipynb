## Geometric

```python
from scipy.stats import geom
```

```python
np.arange(0, 10, 1)
```

```python
p = 0.3

for k in np.arange(0, 10):

    plt.scatter(k, geom.pmf(k, p), label = 'k={0}'.format(k), alpha = 0.8)
plt.legend()
```

```python
pmfs = []
ps = []
ks = []
for k in np.arange(0, 10):
    
    for p in np.linspace(0.01, 0.99, 5):
        ks.append(k)
        ps.append(p)
        pmfs.append(geom.pmf(k, p))
        
    plt.scatter(ps, pmfs, label = 'k={0}'.format(k), alpha = 0.3)
    plt.legend()
```

```python
from scipy.stats import hypergeom
```

```python
x = np.arange(0, 10)
```

```python
for i in x:
    rv = hypergeom(i, 3,4)
    print(rv.pmf(x))
```

```python
def hypergeom_pmf_matrix(number_of_states):
    pmf_matrix = [] 
    x = np.arange(0, number_of_states)
    for i in np.linspace(0, (number_of_states - 1)*2, number_of_states):
        rv = hypergeom((number_of_states - 1)*2, i, number_of_states - 1)
        pmf_vector = rv.pmf(x)
        pmf_vector= pmf_vector
        pmf_matrix.append(pmf_vector)
        
    return pmf_matrix
```
